```python
import numpy as np
import matplotlib.pyplot as plt
from math import *
```

```python
#parameters
nx=100 #no. of points
xmax=10.0 #max length
dx=xmax/(nx-1) #length of subdivision
x0=xmax/2.0 #Point about which gaussian is symmetric
x=np.linspace(0,xmax,nx) #Position definition
a=0.25
f=(1./(sqrt(2*pi*a)))*np.exp(-(x-x0)**2/(2.0*a))
```

```python
plt.figure(figsize=(10,6))
plt.plot(x,f,color="blue",linewidth="2")
plt.xlabel('x')
plt.ylabel('Gaussian function')
plt.xlim(0,xmax)
plt.grid()
plt.show()
```

```python
#2nd order derivative - 3 points
fdashn3=np.zeros(nx)
fdasha=np.zeros(nx)
error=np.zeros(nx)
for i in range(1,nx-1):
    fdashn3[i]=(f[i+1]-2*f[i]+f[i-1])/(dx**2)
fdasha=(1/(sqrt(2*pi*a)))*(np.square(x-x0)/a**2 - 1/a)*np.exp(-(x-x0)**2/(2.0*a))
fdasha[0]=0
fdasha[nx-1]=0
error=fdashn3-fdasha
rms=np.sqrt(np.mean((fdashn3-fdasha)**2))
```

```python
plt.figure(figsize=(10,6))
plt.plot(x,fdashn3,label="Numerical Derivative with 3 points",color="blue",linewidth=2,linestyle="--")
plt.plot(x,fdasha,label="Analytical Derivative",color="red",linewidth=1)
plt.plot(x,error,label="Difference",color="black",linewidth=1,linestyle=":")
plt.xlabel('x')
plt.ylabel('Amplitude')
plt.legend(loc='lower left')
plt.title('Comparision of analytical and numerical derivative with 3 points, rms = %.6f'% (rms))
plt.xlim(0,xmax)
plt.grid()
plt.show()
```

```python
#2nd order derivative with 5 points
fdashn5=np.zeros(nx)
error=np.zeros(nx)
for i in range(2,nx-2):
    fdashn5[i]=((-1/12.0)*f[i-2]+(4/3.0)*f[i-1]-(5/2.0)*f[i]+(4/3.0)*f[i+1]-(1/12.0)*f[i+2])/(dx**2)
fdasha[nx-2]=0
fdasha[1]=0
error=fdashn5-fdasha
rms=np.sqrt(np.mean((fdashn5-fdasha)**2))
```

```python
plt.figure(figsize=(10,6))
plt.plot(x,fdashn5,label="Numerical Derivative with 5 points",color="blue",linewidth=2,linestyle="--")
plt.plot(x,fdasha,label="Analytical Derivative",color="red",linewidth=1)
plt.plot(x,error,label="Difference",color="black",linewidth=1,linestyle=":")
plt.xlabel('x')
plt.ylabel('Amplitude')
plt.legend(loc='lower left')
plt.title('Comparision of analytical and numerical derivative with 5 points, rms = %.6f'% (rms))
plt.xlim(0,xmax)
plt.grid()
plt.show()
```
