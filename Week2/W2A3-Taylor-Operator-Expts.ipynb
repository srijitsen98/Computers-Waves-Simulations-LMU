```python
#import
import numpy as np
from math import *
import matplotlib.pyplot as plt
```

```python
def errorfunc(n1,n2): #n1 is degree n2 is number of points
    n3=(n2-1)/2 
    x=np.zeros(n2)
    A=np.zeros((n2,n2))
    for j in range(0,n2):
        for i in range(0,n2):
            A[i][j]=pow(j-n3,i)
        x[j]=j-n3
    B=np.zeros(n2)
    B[n1]=factorial(n1)/(pow(dx,n1))
    C = np.linalg.solve(A, B) #solving for operator        
```

```python
errorfunc(3,9)
```

```python
plt.figure(figsize=(10,6))
plt.plot(x,C)
plt.title("Taylor operator for degree %d derivative with operator level %d" %(n1,n2))
plt.xlim(-n3,n3)
plt.xlabel('x')
plt.ylabel('Operator')
plt.grid()
plt.show()
```
